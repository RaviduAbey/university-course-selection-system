# Chatgpt's Code with the usage of functions and Classes

```python
class Course:
    def __init__(self, unit_name, unit_code, requirement, semester):
        self.unit_name = unit_name
        self.unit_code = unit_code
        self.requirement = requirement
        self.semester = semester

class Student:
    def __init__(self, st_name, st_id, gender, unit_passed, semester):
        self.st_name = st_name
        self.st_id = st_id
        self.gender = gender
        self.unit_passed = unit_passed
        self.semester = semester

    def get_available_courses(self, courses):
        units_passed = set(self.unit_passed)
        available_courses = []
        for course in courses:
            if course.unit_code not in units_passed:
                if course.requirement is None:
                    if course.semester == self.semester:
                        available_courses.append(course.unit_code)
                elif isinstance(course.requirement, str):
                    if course.requirement in units_passed and course.semester == self.semester:
                        available_courses.append(course.unit_code)
                elif isinstance(course.requirement, list):
                    if set(course.requirement).issubset(units_passed) and course.semester == self.semester:
                        available_courses.append(course.unit_code)
        return available_courses

courses = [
    Course("Coding for BA", "BCO7006", None, 1),
    Course("Biz analytics", "BCO7000", None, 1),
    Course("Predictive Analytics", "BCO6008", "BCO7006", 2),
    Course("Machine Learning", "BCO7007", ["BCO7000", "BCO7006"], 2),
    Course("Women in STEM", "WOM1000", None, "elective")
]

students = [
    Student("Shaini", "221", "F", ["BCO7006", "BCO7000"], 1),
    Student("Kenny", "102", "M", ["BCO7006"], 2),
    Student("Dulini", "318", "F", ["BCO7000"], 1),
    Student("Pathum", "437", "M", ["BCO7007", "BCO6008"], 2),
    Student("Poornima", "438", "F", "", 1)
]

for student in students:
    available_courses = student.get_available_courses(courses)
    if available_courses:
        print(f"Hello {student.st_name}! You can enroll in {', '.join(available_courses)}.")
    else:
        print(f"Hello {student.st_name}! You don't have any subjects to enroll for the upcoming semester.")
```
